```python
ontology = Ontology(
    labels=[
        {"Researcher": "Individuals conducting the research, including principal investigators, co-authors, and contributors."},
        {"Patient": "Subjects of the research studies, including demographic details and medical history."},
        {"Medical Instrument": "Tools and devices used in anaesthetic procedures, such as ventilators, syringes, and monitors."},
        {"Drug": "Pharmaceuticals used in anaesthesia, including names, compositions, and brand names."},
        {"Clinical Trial": "Structured studies to evaluate the effects of anaesthetic interventions, including phases and protocols."},
        {"Surgical Procedure": "Operations and interventions where anaesthesia is administered."},
        {"Anaesthetic Technique": "Specific methods and approaches used in anaesthesia, such as local, regional, and general anaesthesia."},
        {"Symptom": "Signs and symptoms relevant to anaesthesia, such as pain, anxiety, and nausea."},
        {"Condition": "Medical conditions relevant to anaesthesia, including chronic diseases, acute conditions, and allergies."},
        {"Location": "Places where research is conducted, such as hospitals, clinics, and universities."},
        {"Institution": "Medical and research institutions, including universities, hospitals, and research labs."},
        {"Protocol": "Detailed plans and methodologies of anaesthetic procedures and clinical trials."},
        {"Dosage": "Amounts and schedules of drug administration in anaesthesia."},
        {"Adverse Effect": "Negative outcomes and side effects observed in anaesthetic procedures."},
        {"Finding": "Results and conclusions drawn from research studies."},
        {"Funding Source": "Organizations and entities providing financial support for research."},
        {"Ethical Consideration": "Ethical aspects and approvals related to the conduct of research."},
        {"Outcome Measure": "Specific metrics used to evaluate the effectiveness of anaesthetic techniques."},
        {"Tool/Software": "Software or analysis tools used in the research."},
        {"Regulatory Body": "Organizations that regulate the practice and research of anaesthesia."},
        {"Diagnostic Test": "Medical tests used to assess a patient's condition before anaesthesia."},
        {"Biological Sample": "Body fluids or tissues used for analysis."},
        {"Disease": "Specific diseases that are impacted by anaesthetic techniques."},
        {"Diagnostic Test": "Medical tests used to assess a patient's condition before anaesthesia."},
        {"Miscellaneous": "Any other significant concepts not categorized under the provided labels."},
        ],
    
    relationships=[
        "Conducted by",  # Relation between a Clinical Trial and a Researcher or Institution conducting it.
        "Administers",  # Relation between a Researcher and a Drug or Dosage they administer.
        "Located at",  # Relation between an Event or Institution and its Location.
        "Supported by",  # Relation between a Research Project and its Funding Source.
        "Uses",  # Relation between a Procedure or Technique and the Medical Instrument or Drug used.
        "Causes",  # Relation between a Drug or Procedure and an Adverse Effect or Symptom it causes.
        "Treats",  # Relation between a Drug or Procedure and the Condition or Symptom it treats.
        "Results in",  # Relation between a Procedure or Clinical Trial and its Finding or Outcome.
        "Follows",  # Relation between a Procedure and the Protocol it follows.
        "Ethically Approved by",  # Relation between a Clinical Trial or Procedure and the Ethical Consideration or board approving it.
        "Authored by",  # Relation between a Publication and its Researcher or Author.
        "Affiliated with",  # Relation between a Researcher and their Institution.
        "Observed in",  # Relation between an Adverse Effect or Symptom and the Patient or Clinical Trial where it was observed.
        "Described in",  # Relation between a Finding and the Publication describing it.
        "Diagnosed by", #Between a Patient and Diagnostic Test used to assess their condition.
        "Related to",  # General relationship between any pair of Entities not covered by specific relationships.
        "Complies with", #Between a Protocol or Clinical Trial and Guidelines or Regulations.
        "Extracted from", #Between Biological Samples and the Patient they are obtained from.,
    ]
)
```

```python
import os
from transformers import GPT2Tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Function to split text into chunks of desired token size
def split_into_chunks(text, chunk_size=1000):
    tokens = tokenizer.encode(text)
    chunks = [tokens[i:i + chunk_size] for i in range(0, len(tokens), chunk_size)]
    return [tokenizer.decode(chunk) for chunk in chunks]

# Function to read all text files from a folder
def read_text_files_from_folder(folder_path):
    combined_text = ""
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                combined_text += file.read() + " "  # Add a space between file contents
    return combined_text

# Folder path containing the text files
folder_path = "Downloads/papers_text"


# Read text files and create the combined text
text = read_text_files_from_folder(folder_path)

# Split the combined text into chunks
chunks = split_into_chunks(text)

# Convert chunks into Document instances
documents = [Document(text=chunk, metadata={"source": "example"}) for chunk in chunks]

# Print the documents
for doc in documents:
    print(doc)
```

```python
from knowledge_graph_maker import OpenAIClient

model = "gpt-4"

os.environ['OPENAI_API_KEY'] = YOUR_API_KEY
## -> Initiate the Groq Client.

llm = OpenAIClient(model=model, temperature=0.1, top_p=0.5)
graph_maker = GraphMaker(ontology=ontology, llm_client=llm, verbose=False)

## -> Create a graph out of a list of Documents.
graph = graph_maker.from_documents(documents)
## result: a list of Edges.

print("Total number of Edges", len(graph))
```

```python
import pandas as pd

edges_data = []
for edge in graph:
    edges_data.append({
        "source": edge.node_1.name,
        "target": edge.node_2.name,
        "relationship": edge.relationship
    })

edges_df = pd.DataFrame(edges_data)

edges_df
```

```python
from neo4j import GraphDatabase

NEO4J_URI= YOUR_NEO4J_URI
NEO4J_USERNAME= YOUR_NEO4J_USERNAME
NEO4J_PASSWORD= YOUR_NEO4J_PASSWORD
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))

# Here is the modified function to create nodes of different labels and connect them with a 'RELATED_TO' relationship
def add_relations(tx, source, target, description):
    tx.run("MERGE (a:Entity {name: $source}) "
           "MERGE (b:Entity {name: $target}) "
           "MERGE (a)-[r:RELATED_TO {description: $description}]->(b)",
           source=source, target=target, description=description)

# Use the driver session to write the data from your dataframe to the Neo4j database
with driver.session() as session:
    for index, row in edges_df.iterrows():
        session.write_transaction(add_relations, row['source'], row['target'], row['relationship'])

# Close the driver session once done
driver.close()
```
